Read in the abalone data. You can retrieve the link from this URL: [abalone data folder](https://archive.ics.uci.edu/ml/machine-learning-databases/abalone/)

Use the `col_names` argument with the `names` vector from above to set the column names.

```r
names = c('Length', 'Diameter', 'Height', 'Whole_weight', 'Shucked_weight', 'Viscera_weight', 'Shell_weight', 'Rings' )
```

```r
names
```

```r
dfabalone = read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/abalone/abalone.data',
         col_names = names,
         col_types = cols())
         
dfabalone  %>% head
```

```r
abalone_df = read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/abalone/abalone.data',
        col_type = cols(),
        col_names = names)

abalone_df  %>% head
```

Is there any missing data? Use `filter` to find out.

```r
abalone_df %>% filter(if_any(everything(), is.na))
#this shows there are no missing values

#another way to do this is to put:
    #
```
