```r
options(repr.plot.width=6, repr.plot.height=2.5, repr.plot.res = 300)

# write your chart code below ðŸ‘‡
p = ggplot(cases, aes(x=date, y=positiveIncrease))
p = p + geom_col(width = 1)
#if you do not add width = 1 in geom_col, the width of the bars will not meet side by side in terms of fill, this is 
    #to give the user more clarity in terms of counts for each date, but to make the chart in the given picture, we must
    #use width = 1
p
```

# Same chart as above, but using scales package for commas

```r
options(repr.plot.width=6, repr.plot.height=2.5, repr.plot.res = 300)

# write your chart code below ðŸ‘‡
p = ggplot(cases, aes(x=date, y=positiveIncrease))
p = p + geom_col(width = 1)
p = p + scale_y_continuous(labels = comma)
#uses package to bring in commas for formatting purposes
p
```
