Use `filter` and `summarize` to produce two single row tables. One that has the median `temp` for January, and one that has the median `temp` for June.

```r
# E.g.
m = 1

 median_temp  = weather %>% 
     #group_by(month) %>%
     filter(month == m) %>% 
     summarize(median_temp = median(temp))

month_temp = weather  %>% 
    filter(month == m)



p = ggplot(month_temp, aes(x = day, y = temp, color = temp))

p = p + geom_point(position = position_jitter(h=0, w= 0.2), size =1)

p = p + geom_hline(data = median_temp, aes(yintercept = median_temp))

p = p + scale_colour_viridis_c()

p = p + ylim(5,100)

p_january = p

p_january
```

```r
# E.g.
m = 6

 median_temp  = weather %>% 
     #group_by(month) %>%
     filter(month == m) %>% 
     summarize(median_temp = median(temp))

month_temp = weather  %>% 
    filter(month == m)



p = ggplot(month_temp, aes(x = day, y = temp, color = temp))

p = p + geom_point(position = position_jitter(h=0, w= 0.2), size =1)

p = p + geom_hline(data = median_temp, aes(yintercept = median_temp))

p = p + scale_colour_viridis_c()

p = p + ylim(5,100)

p_june = p

p_june
```

```r
p_january + p_june
#We can do this because of patchwork
```
