Start with a plot object


````r
p = ggplot(mpg, aes(x = cty, y = hwy))
#p is just a random variable to reference the output from all of the instructions from ggplot

#first argument is the data and the second is the encodings for the variables to be titled
    #mapping column names to chart properties

#automatically defualt to scatter plot?
````

````r
p
#this is just the initial set up of the table, but we need to add the data (add layer onto the outline of the chart 
#created)
````

Add the `geom` you want to represent the data.

````r
p2 = p + geom_point(size = 3)
#takes the outline produced above and adds the data points through adding geom_point(size = 3) to the p we created

#Another convention could be to write the code this way:
    #p = ggplot(mpg, aes(x = cty, y = hwy)) + geom_point(size = 3)
    #this provides the same output in one step

p2
````

````r
#Adding stylistic changes to the plot we have created above
p = ggplot(mpg, aes(x = cty, y = hwy))
p2 = p + geom_point(size = 5, color = 'blue')
p2
````

What if we want to color the points by `class`?

````r
p = ggplot(mpg, aes(x = cty, y = hwy)) +
    geom_point(size = 2)

p

#Want to add different colors to points depending on the class variable they are designated by
#also change the size based on cty
p2 = ggplot(mpg, aes(x = cty, y = hwy, color = class, size = cty)) +
    geom_point()
#cannot define size in two spots or the second will overwrite the first in the outline

p2
#add color = class and size = cty statement within the mapping for the variables in the outline
````
