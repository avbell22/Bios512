Use `arrange` to find the best race time.

```r
df.marathon  %>% 
    arrange(race.time)  %>% 
    head(1)
```

What are the best race times by country?

```r
df.marathon  %>% 
    group_by(country) %>% 
    summarise(best.time = min(race.time))  %>% 
    arrange(best.time)

#Other Examples
    
    #df.marathon  %>% 
        #group_by(country)  %>% 
        #arrange(race.time)  %>% 
        #mutate (rk = row_number())  %>% 
        #filter(rk == 1)
    
    #df.marathon  %>% 
        #group_by(country) %>% 
        #mutate (rk = rank(desc(race.time)))  %>% 
        #filter (rk == 1)  %>% 
        #arrange(race.time)
    
```

In what year was the difference between the men's and women's winning time smallest? Use this table:

```r
 df = df.marathon %>%
     select(year, MF, race.time) %>%
     tidyr::spread(MF, race.time) %>%
     drop_na() %>%
    #need the drop function to get rid of rows where there are missing values so we can do functions on data
        #this means that we are missing the years where there are no women's race times
     rename(men = `Men's`, women = `Women's`)
    #this allows us to call these variables easier
 df %>% head
```

```r
df  %>% 
    mutate(diff_mw_time = abs(men - women))  %>% 
    arrange (diff_mw_time)  %>% 
    head(1)
```

```r
time = df  %>% 
         mutate(diff_mw_time = abs(men - women))  %>% 
         arrange (diff_mw_time)  
```

Plot the difference between men's and women's race times by year.

```r
 options(repr.plot.res = 250, repr.plot.width = 5, repr.plot.height = 4)
```

```r
p = ggplot(time, aes(x = year, y = diff_mw_time))
p = p + geom_line()
p
```
