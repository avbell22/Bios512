```r
mpg_longer =
mpg  %>% 
    mutate(id = row_number())  %>% 
    pivot_longer(c(cty, hwy), names_to = 'label', values_to = 'TEST')

mpg_longer
```

```r
#This is all for debugging
df = mpg_longer  %>% filter(id == 1)

p = ggplot(df, aes(x = TEST, y= id))

p = p + geom_point()

p
```

```r
top_10 = mpg_longer  %>% 
    pivot_wider(names_from = 'label', values_from = TEST)  %>% 
    slice_max(cty, n =10)  %>% 
    pull(id)



```

```r
df_filtered = mpg_longer  %>% 
    filter(id %in% top_10)  %>% 
    mutate(id = fct_reorder(as.character(id), TEST, max))
```

```r
options(repr.plot.height = 2)
p = ggplot(df_filtered, aes(x = TEST, y = as.character(id)))

p = p + geom_point()

p
```

```r

df_filtered = mpg_longer %>%
    filter(id %in% top_10) %>%
    mutate(id = fct_reorder(as.character(id), TEST, max))

mpg_wider = mpg_longer %>%
    pivot_wider(names_from = label, values_from = TEST) %>%
    slice_max(cty, n = 10) %>%
    mutate(id = as.character(id))

p = ggplot(df_filtered, aes(x = TEST, y = id))

p = p + geom_point(mapping = aes(color = label))

p = p + geom_segment(data = mpg_wider, 
                     mapping = aes(x = cty, xend = hwy, y = id, yend = id))

p
```
